```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0 , 10 , 20 , 40 , 40 , 58 , 60])
y = np.array([0 , 8 , 6 , 26 , 24 , 51 , 60])

plt.plot(x , y , color = "g")

plt.title("Daily stock exchanges")
plt.xlabel("no.of products available")
plt.ylabel("remaining stocks")

plt.show()
```
